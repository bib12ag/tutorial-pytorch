### Transform
*module*
- `torchvision.transforms`
    - 데이터를 학습에 알맞은 형태로 변형시키기 위한 모듈

*method*
- `Resize`
    - 이미지 크기 변경
- `ToTensor`
    - 이미지를 pytorch tensor 타입으로 변형하여 픽셀 값을 [0, 255]에서 [0.0, 1.0]으로 자동 변환
- `Compose`
    - 여러 변형을 custom하여 한 번에 수행

```python
import torch 
from torchvision import transforms, datasets
from torchvision.transforms import Resize, ToTensor, Compose
import matplotlib.pyplot as plt 
```

```python
transform = Compose([Resize((32,32)), ToTensor()])

dataset = datasets.MNIST(root='.', download=True)
trans_dataset = datasets.MNIST(root='.', download=True, transform=transform)

img = dataset[0][0]
trans_img = transforms.ToPILImage()(trans_dataset[0][0])

plt.subplot(1,2,1)
plt.title('28x28 img')
plt.imshow(img, cmap='gray')

plt.subplot(1,2,2)
plt.title('32x32 transformed img')
plt.imshow(trans_img, cmap='gray')
```
